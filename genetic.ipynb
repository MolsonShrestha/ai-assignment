```python


import random
import math

#Constants
a = 1
b = 9
c = 18

#Genetic algorithm parameters
POPULATION_SIZE = 5
NUM_GENERATIONS = 5
MUTATION_RATE = 0.1

# Binary encoding
SIGN_BIT = 1
INTEGER_BITS = 3
FRACTIONAL_BITS = 4
CHROMOSOME_LENGTH = SIGN_BIT + INTEGER_BITS + FRACTIONAL_BITS

# Calculating the fitness
def fitness_function(x):
    # Decode the binary string to decimal
    decimal_x = decode_binary(x)
    
    # Calculate the fitness based on the equation
    fitness = abs(a * decimal_x**2 + b * decimal_x + c)
    
    return fitness

# Decoding
def decode_binary(binary_string):
    # Extract sign bit, integer part, and fractional part
    sign = int(binary_string[0])
    integer_part = int(binary_string[1:1+INTEGER_BITS], 2)
    fractional_part = int(binary_string[1+INTEGER_BITS:], 2) / (2 ** FRACTIONAL_BITS)
    
    decimal_value = integer_part + fractional_part
    

    if sign == 1:
        decimal_value = -decimal_value
    
    return decimal_value

# Initializing population
def initialize_population():
    population = []
    for _ in range(POPULATION_SIZE):
        # Generate random binary string
        binary_string = ''.join([str(random.randint(0, 1)) for _ in range(CHROMOSOME_LENGTH)])
        population.append(binary_string)
    return population

# Perform uniform crossover
def uniform_crossover(parent1, parent2):
    child = []
    for i in range(len(parent1)):
        if random.random() < 0.5:
            child.append(parent1[i])
        else:
            child.append(parent2[i])
    return ''.join(child)

# Performing mutation
def mutate(individual):
    mutated_individual = list(individual)
    for i in range(len(mutated_individual)):
        if random.random() < MUTATION_RATE:
            mutated_individual[i] = '0' if mutated_individual[i] == '1' else '1'
    return ''.join(mutated_individual)

#genetic algorithm function
def genetic_algorithm():
    population = initialize_population()
    best_solution = None
    best_fitness = float('inf')
    
    for generation in range(NUM_GENERATIONS):
        # Evaluating fitness of each individual
        fitness_scores = []
        for individual in population:
            fitness_score = fitness_function(individual)
            fitness_scores.append((individual, fitness_score))
            fitness_scores.sort(key=lambda x: x[1])
        
        #best fitness
        current_best_fitness = fitness_scores[0][1]
        if current_best_fitness < best_fitness:
            best_fitness = current_best_fitness
            best_solution = fitness_scores[0][0]
        print(f"Generation {generation}: Best Fitness = {best_fitness}")
        
        # Selecting top individuals for reproduction
        selected_parents = [individual for individual, _ in fitness_scores[:2]]
        
        #crossover and mutation
        next_generation = []
        while len(next_generation) < POPULATION_SIZE:
            parent1 = random.choice(selected_parents)
            parent2 = random.choice(selected_parents)
            
            child = uniform_crossover(parent1, parent2)
            mutated_child = mutate(child)
            
            next_generation.append(mutated_child)
        
        # Update population with the new generation
        population = next_generation
    
    # Decode the best solution found
    decoded_best_solution = decode_binary(best_solution)
    
    return decoded_best_solution

# Running genetic algorithm
best_solution = genetic_algorithm()
print(f"Best solution found: x = {best_solution}")

# printing
discriminant = b**2 - 4*a*c
if discriminant >= 0:
    root1 = (-b + math.sqrt(discriminant)) / (2*a)
    root2 = (-b - math.sqrt(discriminant)) / (2*a)
    print(f"Roots of the equation are: x1 = {root1}, x2 = {root2}")
else:
    print("No real roots exist for the given quadratic equation.")
```
